```python
import pandas as pd
```

```python
bank_as='explore_bank.csv'
bank=pd.read_csv(bank_as, sep=';', decimal='.')
```

```python
bank
```

```python
bank=bank.drop(bank.columns[0], axis=1)
```

```python
bank.loc[2]
```

```python
import matplotlib.pyplot as plt
```

```python
#The attribute "job" is plotted in a pie chart to depict the various catagories.
bank['job'].value_counts().plot(kind='pie',autopct='%.2f',legend=True)
plt.legend(bbox_to_anchor=(1.5,1.0))
```

```python
#The attribute "job" is plotted in a histogram chart to depict the various catagories of age and height
bank['age'].plot(kind='hist',bins=20,legend=True,color='red')
plt.title('Age')
plt.xlabel('Height')
```

```python
#The attribute "marital" is plotted in a pie chart to depict the various catagories such as married,single,divorced and unknown

bank['marital'].value_counts().plot(kind='pie',autopct='%.2f',legend=True)
```

```python
#The attribute "education" is plotted in apie chart to depict the various catagories.
bank['education'].value_counts().plot(kind='pie',autopct='%.2f',legend=True)
plt.legend(bbox_to_anchor=(1.5,1.0))
```

```python
#The attribute "default" is plotted in a bar chart to depict the various catagories of default and frequency.
bank['default'].value_counts().plot(kind='bar')
plt.title('Default')
plt.ylabel('Frequency')
```

```python
#The attribute "Housing" is plotted in a bar chart to depict the various catagories such as yes,no and unknown.
ax=bank['housing'].value_counts().plot(kind='bar',color=['black','orange','red'],legend=True)
plt.title('Housing')
ax.legend(['yes','no','unknown'])

plt.ylabel('Frequency')
```
